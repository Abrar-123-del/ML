```python
import pandas as pd
```

```python
df = pd.read_csv("heart.csv")
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_Sex = LabelEncoder()
le_ChestPainType = LabelEncoder()
le_RestingECG = LabelEncoder()
le_ExerciseAngina = LabelEncoder()
le_ST_Slope = LabelEncoder()
```

```python
df['Sex_n'] = le_Sex.fit_transform(df['Sex'])
df['ChestPainType_n'] = le_ChestPainType.fit_transform(df['ChestPainType'])
df['RestingECG_n'] = le_RestingECG.fit_transform(df['RestingECG'])
df['ExerciseAngina_n'] = le_ExerciseAngina.fit_transform(df['ExerciseAngina'])
df['ST_Slope_n'] = le_ST_Slope.fit_transform(df['ST_Slope'])
```

```python
df.head()
```

```python
df_n = df.drop(['Sex', 'ChestPainType', 'RestingECG', 'ExerciseAngina', 'ST_Slope'], axis='columns')
```

```python
df_n
```

```python
y = df_n['HeartDisease']
y
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(0.95)
```

```python
X_pca = pca.fit_transform(df_n.drop('HeartDisease', axis='columns'))
```

```python
X_pca.shape
```

```python
X_pca
```

```python
pca.explained_variance_ratio_
```

```python
pca.n_components_
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=2)
```

```python
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
model_1 = SVC()
model_1.fit(X_train, y_train)
model_1.score(X_test, y_test)
```

```python
model_2 = LogisticRegression()
model_2.fit(X_train, y_train)
model_2.score(X_test, y_test)
```

```python
model_3 = RandomForestClassifier()
model_3.fit(X_train, y_train)
model_3.score(X_test, y_test)
```
